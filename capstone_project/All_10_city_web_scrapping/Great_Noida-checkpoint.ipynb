```python
#fetching urls
Url_6=[]
for page in range(0,12):
    page_url=driver.find_elements(By.XPATH,'//a[@class="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"]')
    for i in page_url:
        Url_6.append(i.get_attribute('href'))
```

```python
BHK_types=[]
for page in range(0,12):
    try:
        bhk=driver.find_elements(By.XPATH,'//div[@class="font-semibold whitespace-nowrap"]')
        for bhk_element in bhk:
            BHK_types.append(bhk_element.text)
    except NoSuchElementException:
        print("-")
```

```python
Carpet_area=[]
for page in range(0,12):
    try:
        carpet=driver.find_elements(By.XPATH,'//div[@class="font-semi-bold heading-6"]/div')
        for i in carpet:
            Carpet_area.append(i.text)
    except NoSuchElementException:
        print("-")
    
```

```python
#fetching buy carpet area
Price_per_sq=[]
for page in range(0,12):
    
      try:
        price=driver.find_elements(By.XPATH,'//div[@class="nb__7nqQI"]/div[1]/div[2]/div[2]')
           
        for j in price:
               Price_per_sq.append(j.text)
      except NoSuchElementException:
        print("-")
```

```python
Car_parking = []

for page in range(0, 312):
    
    try:
        # Fetch car parking element
        car_park = driver.find_element(By.XPATH, '//div[@class="border-3 border-double border-property-views-border-color flex flex-col"]/div[2]/div[2]/div[2]/div[1]')

            
        # Normalize text and check for parking availability
        car_park_text = car_park.text.strip().lower()
        category = 'yes' if 'bike and car' in car_park_text or 'car' in car_park_text else 'no'
        
        # Append category to Car_parking list
        Car_parking.append(category)
        
    except NoSuchElementException:
        Car_parking.append('No')  # Append 'No' if element not found or handle as needed

Car_parking
```

```python
# Initialize lists to store data
Beds = []
Bathroom = []
Balcony = []
Address = []
City=[]
State_code=[]
Country=[]
Landmarks=[]
Furnished_status=[]
Facing=[]
Type_of_ownership=[]
booking_amt=[]
buy_total_price=[]
Status=[]
Floor=[]
Transition_Type=[]
year_built=[]


# Iterate through each URL to fetch data
for url in Url_6:
    driver.get(url)
    
    ## Fetch bed data
    try:
        bed = driver.find_element(By.XPATH, '//*[@id="details-summary-typeDesc"]')
        Beds.append(bed.text)
    except NoSuchElementException:
        Beds.append('-')

    ## Fetch bathroom data
    try:
        bathroom = driver.find_element(By.XPATH, '//h4[@id="details-summary-bathroom"]')
        Bathroom.append(bathroom.text)
    except NoSuchElementException:
        Bathroom.append('-')

    ## Fetch Balcony data
    try:
        balcony = driver.find_element(By.XPATH, '//div[@class="nb__3Z_gh"][5]/div[2]/span')
        Balcony.append(balcony.text)
    except NoSuchElementException:
        Balcony.append('-')

    ## Fetch address of the house
    try:
        add = driver.find_element(By.XPATH, '//h5[@class="text-muted nb__16pur"]')
        Address.append(add.text)
    except NoSuchElementException:
        Address.append('-')

    ##fetching city
    try:
        City.append('Great_Noida')

    except NouSuchElementException:
        City.append('-')

   
    ##fetching state code
    try:
        State_code.append('37')
    except NouSuchElementException:
        State_code.append('-')

    
     ##fetching country
    try:
        Country.append('India')

    except NouSuchElementException:
        Country.append('-')


     ##fetching Landmarks
    try:
        land=driver.find_element(By.XPATH,'//div[@class="cursor-pointer nb__1Fvam"]/p')
        Landmarks.append(land.text)
    except NoSuchElementException:
        Landmarks.append('-')


    ##fetching Furnished status
    try:
        Furnished=driver.find_element(By.XPATH,'//div[@class="nb__3ocPe"][6]/div[3]')
        Furnished_status.append(Furnished.text)
    except NoSuchElementException:
        Furnished_status.append('-')


    ##fetching facing
    try:
        face=driver.find_element(By.XPATH,'//*[@id="property-detail-body"]/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[7]/div[3]/h5')
        Facing.append(face.text)
    except NoSuchElementException:
        Facing.append('-')

    

   ##fetching Type of ownership
    try:
        ownership=driver.find_element(By.XPATH,'//div[@class="nb__28cwR"]/div[2]/div[3]')
        Type_of_ownership.append(ownership.text)
    except NoSuchElementException:
        Type_of_ownership.append('-')


    #fetching booking amount
    #try:
        #book_amt=driver.find_element(By.XPATH,'')
        #booking_amt.append(book_amt.text)
    #except NoSuchElementException:
        #booking_amt.append('-')


    ##fetching buy total price
    try:
        buy_price=driver.find_element(By.XPATH,'//div[@id="formattedPrice"]')
        buy_total_price.append(buy_price.text)
    except NoSuchElementException:
        buy_total_price.append('-')


         
    ##fetching status
    try:
        
        Status.append('complete')
    except NoSuchElementException:
        Status.append('-')


    ##fetching Floor
    try:
        floor=driver.find_element(By.XPATH,'//*[@id="property-detail-body"]/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[8]/div[3]/h5')
        Floor .append(floor.text)
    except NoSuchElementException:
        Floor.append('-')

   
    ##fetching Transition type#  
    try:
        Transition=driver.find_element(By.XPATH,'//div[@class="nb__3Cqpf"]')
        Transition_Type .append(Transition.text)
    except NoSuchElementException:
        Transition_Type.append('-')
    Transition_Type

   
    
    ##fetching year built
    try:
         year=driver.find_element(By.XPATH,'//*[@id="property-detail-body"]/div/div[1]/div/div/section[2]/div/div[1]/div[1]/div/div[2]/div[1]/div[3]/h5')
         year_built .append(year.text)
    except NoSuchElementException:

    ##year_built
        year_built.append('-')
```

```python
property_types = []
for url in Url_6:
    try:
        driver.get(url)
        property_type_element = driver.find_element(By.XPATH, '//div[@class="nb__3Z_gh"][6]/div[2]/h5[1]')
        property_type = property_type_element.text.strip().lower()
        category = 'condo' if 'apartment' in property_type else 'house'
        property_types.append(category)

    except NoSuchElementException:
        property_types.append('-')
```
